# Making an Interactive Map of MSOE

## Initializing The Map


```python
import folium

# Where map centers
# folium only works with latitude and longitude coordinates
msoe_center = [43.0448, -87.9065]

# Creates the map
msoe_map = folium.Map(location=msoe_center, zoom_start=17)
```

## Adding Markers (2 Different Ways)

### Process 1: Manual Addition
Works for adding a few locations, but gets repetitive for maps requiring multiple markers.

```python
# each marker is added individually
folium.Marker([43.043, -87.9065], popup="Science Hall").add_to(msoe_map)
folium.Marker([43.0443, -87.907628], popup="Diercks Hall").add_to(msoe_map)
folium.Marker([43.046501, -87.909271], popup="Kern Center").add_to(msoe_map)
folium.Marker([43.043808, -87.907938], popup="Grohmann Museum").add_to(msoe_map)
folium.Marker([43.045319, -87.907707], popup="Viets Tower").add_to(msoe_map)
folium.Marker([43.045513, -87.910464], popup="Grohmann Tower").add_to(msoe_map)
folium.Marker([43.043923, -87.909044], popup="Campus Center").add_to(msoe_map)
folium.Marker([43.044203, -87.906445], popup="Raider Field").add_to(msoe_map)
folium.Marker([43.047713, -87.909264], popup="Viets Field").add_to(msoe_map)
folium.Marker([43.042906, -87.905098], popup="Walter Schroder Library").add_to(msoe_map)
folium.Marker([43.046151, -87.907670], popup="Margaret Loock").add_to(msoe_map)
folium.Marker([43.045552, -87.907659], popup="Mellowes").add_to(msoe_map)
folium.Marker([43.046635, -87.907678], popup="Rosenberg Hall").add_to(msoe_map)
msoe_map
```

### Process 2: Loop Addition
Don't need to rely on individually adding every marker. 

```python
# list of tuples with the necessary information
locations = [
    ("Science Hall", 43.043, -87.9065),
    ("Diercks Hall", 43.0443, -87.907628),
    ("Kern Center", 43.046501, -87.909271),
    ("Grohmann Museum", 43.043808, -87.907938),
    ("Viets Tower", 43.045319, -87.907707),
    ("Grohmann Tower", 43.045513, -87.910464),
    ("Campus Center", 43.043923, -87.909044),
    ("Raider Field", 43.044203, -87.906445),
    ("Viets Field", 43.047713, -87.909264),
    ("Walter Schroder Library", 43.042906,-87.905098),
    ("Margaret Loock", 43.046151, -87.907670),
    ("Mellowes", 43.045552, -87.907659),
    ("Rosenberg Hall", 43.046635, -87.907678)
]
# loop iterates through the indexes of the tuples
for area in locations:
    folium.Marker(
        location=[area[1], area[2]],
        popup = area[0]
).add_to(msoe_map)
msoe_map
        
```

## Map Customization

```python
# Folium sometimes generates previous edits from memory, helpful to re-initialize the map
msoe_map2 = folium.Map(location=msoe_center, zoom_start=17)
# a 4th 'type' entry in the tuple list has been added
locations_custom = [
    ("Science Hall", 43.043, -87.9065,'Class'),
    ("Diercks Hall", 43.0443, -87.907628,'Class'),
    ("Kern Center", 43.046501, -87.909271,'Sports'),
    ("Grohmann Museum", 43.043808, -87.907938,'Class'),
    ("Viets Tower", 43.045319, -87.907707,'Student Life'),
    ("Grohmann Tower", 43.045513, -87.910464,'Student Life'),
    ("Campus Center", 43.043923, -87.909044,'Student Life'),
    ("Raider Field", 43.044203, -87.906445,'Sports'),
    ("Viets Field", 43.047713, -87.909264, 'Sports'),
    ("Walter Schroder Library", 43.042906,-87.905098,'Class'),
    ("Margaret Loock", 43.046151, -87.907670,'Student Life'),
    ("Mellowes", 43.045552, -87.907659,'Student Life'),
    ("Rosenberg Hall", 43.046635, -87.907678,'Class')
]
# determines what the marker should look like based on type
for loc in locations_custom:
    if 'Class' in loc[3]:
        mark_color = 'red'
        mark = 'book'
    elif 'Sports' in loc[3]:
        mark_color = 'green'
        mark = 'star'
    elif 'Student Life' in loc[3]:
        mark_color = 'blue'
        mark = 'home'
# adds the markers into the map
    folium.Marker(
        location = [loc[1],loc[2]],
        tooltip = loc[0],
        icon = folium.Icon(color = mark_color, icon = mark)).add_to(msoe_map2)
# Defines the desired boundary
campus_boundary = [
    [43.0485, -87.9110], 
    [43.0485, -87.9045], 
    [43.0425, -87.9045], 
    [43.0425, -87.9110] 
]
# .Polygon draws a line from point-to-point
folium.Polygon(
    locations=campus_boundary,
    color="blue",  
    fill = False,
    popup="MSOE Campus"
).add_to(msoe_map2)
msoe_map2
```

## Data Distribution
Folium allows you to not only add markers, but also visualize values onto a map

```python
# gives you the outline of each US state
import pandas as pd
import requests
import folium
states = requests.get(
    "https://raw.githubusercontent.com/python-visualization/folium-example-data/main/us_states.json"
).json()

country = folium.Map(location=[48, -102], zoom_start=3)

# dictionary containing the MSOE student population
msoe_pop = {
    'Wisconsin': 326, 'Illinois': 118, 'Michigan': 9, 'California': 8, 
    'Texas': 7, 'Colorado': 4, 'Florida': 3, 'North Carolina': 3, 
    'New York': 3, 'Indiana': 3, 'Iowa': 2, 'South Dakota': 2, 
    'Pennsylvania': 2, 'Ohio': 2, 'Arizona': 2, 'Kentucky': 1, 
    'New Jersey': 1, 'Idaho': 1, 'Hawaii': 1, 'Oregon': 1, 
    'Georgia': 1, 'District of Columbia': 1, 'Tennessee': 1, 
    'Connecticut': 1, 'Virginia': 1, 'Massachusetts': 1
}

# panda cannot handle raw lists, must translate it using .Dataframe
panda_data = pd.DataFrame(list(msoe_pop.items()), columns=['State','Students'])


folium.Choropleth(
    geo_data = states,
    data = panda_data,
    columns=['State','Students'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrRd',
    bins = [0,5,100,200,330],
    fill_opacity = '.7',
    line_opacity = '.2',
    legend_name ='Students Per State').add_to(country)
country
    
```
