# Making an Interactive Map of MSOE

## Initializing The Map


```python
import folium

# Center of MSOE campus
msoe_center = [43.0448, -87.9065]

# Create map
msoe_map = folium.Map(location=msoe_center, zoom_start=17)
```

## Adding Markers (2 Different Ways)

### Process 1: Direct Addition

```python
folium.Marker([43.043, -87.9065], popup="Science Hall").add_to(msoe_map)
folium.Marker([43.0443, -87.907628], popup="Diercks Hall").add_to(msoe_map)
folium.Marker([43.046501, -87.909271], popup="Kern Center").add_to(msoe_map)
folium.Marker([43.043808, -87.907938], popup="Grohmann Museum").add_to(msoe_map)
folium.Marker([43.045319, -87.907707], popup="Viets Tower").add_to(msoe_map)
folium.Marker([43.045513, -87.910464], popup="Grohmann Tower").add_to(msoe_map)
folium.Marker([43.043923, -87.909044], popup="Campus Center").add_to(msoe_map)
folium.Marker([43.044203, -87.906445], popup="Raider Field").add_to(msoe_map)
folium.Marker([43.047713, -87.909264], popup="Viets Field").add_to(msoe_map)
folium.Marker([43.042906, -87.905098], popup="Walter Schroder Library").add_to(msoe_map)
folium.Marker([43.046151, -87.907670], popup="Margaret Loock").add_to(msoe_map)
folium.Marker([43.045552, -87.907659], popup="Mellowes").add_to(msoe_map)
folium.Marker([43.046635, -87.907678], popup="Rosenberg Hall").add_to(msoe_map)
msoe_map
```

### Process 2: Loop Addition

```python
locations = [
    ("Science Hall", 43.043, -87.9065),
    ("Diercks Hall", 43.0443, -87.907628),
    ("Kern Center", 43.046501, -87.909271),
    ("Grohmann Museum", 43.043808, -87.907938),
    ("Viets Tower", 43.045319, -87.907707),
    ("Grohmann Tower", 43.045513, -87.910464),
    ("Campus Center", 43.043923, -87.909044),
    ("Raider Field", 43.044203, -87.906445),
    ("Viets Field", 43.047713, -87.909264),
    ("Walter Schroder Library", 43.042906,-87.905098),
    ("Margaret Loock", 43.046151, -87.907670),
    ("Mellowes", 43.045552, -87.907659),
    ("Rosenberg Hall", 43.046635, -87.907678)
]
for area in locations:
    folium.Marker(
        location=[area[1], area[2]],
        popup = area[0]
).add_to(msoe_map)
msoe_map
        
```
